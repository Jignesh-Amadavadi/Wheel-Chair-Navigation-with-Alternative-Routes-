```python
from Service.Navigation import *
from time import time
```

```python
class Request:
    def __init__(self, lat1, lon1, lat2, lon2, isWheelChair, isShortestPath, distance_weight, epsilon, gamma, alpha):
        self.lat1 = lat1
        self.lon1 = lon1
        self.lat2 = lat2
        self.lon2 = lon2
        self.isWheelChair = isWheelChair
        self.isShortestPath = isShortestPath
        self.distance_weight = distance_weight
        self.epsilon = epsilon
        self.gamma = gamma
        self.alpha = alpha
```

```python
req = Request(
    lat1=50.7359851, lon1=7.1012430,
    lat2=50.7283926, lon2=7.0699449,
    isWheelChair=False,
    isShortestPath=True,
    distance_weight="length",  # or "weight" for wheelchair
    epsilon=0.25,
    gamma=0.5,
    alpha=0.3
)
```

```python
start = time()
res = shortest_path_by_walking(req)
shorts_time = time() - start
```

```python
start = time()
res = alternative_paths_by_walking(req)
alternative_time = time() - start
```

```python
print(f"Shortest path time: {shorts_time:.3f}s")
print(f"Alternative path time: {alternative_time:.3f}s")
```
